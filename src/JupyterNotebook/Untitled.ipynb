```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return (np.sin(x)/2 + np.cos(x)/2)**2 * np.sin(x)
```

```python
roots = np.roots([1, 0, 0, 1, 0])
print("Корни функции f(x) =", roots)
```

```python
x = np.linspace(0, 2*np.pi, 1000)
y = f(x)
deriv = np.diff(y)
increasing_intervals = np.where(deriv > 0)
print("Интервалы возрастания функции:", x[increasing_intervals])
```

```python
decreasing_intervals = np.where(deriv < 0)
print("Интервалы убывания функции:", x[decreasing_intervals])
```

```python
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('График функции f(x)')
plt.show()
```

```python
vertex_index = np.argmax(y)
vertex_value = y[vertex_index]
vertex = x[vertex_index]
print("Вершина функции f(x) находится в точке x =", vertex, " с f(x) =", vertex_value)
```
